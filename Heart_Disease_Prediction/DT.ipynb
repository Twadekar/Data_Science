```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the Dataset
df = pd.read_excel("heart_disease.xlsx", sheet_name="Heart_disease")
```

```python
df.head()
```

```python
#Check Null Values
df.isnull().sum()
```

```python
df.shape
```

```python
#Fill null values with mean
df["oldpeak"] = df["oldpeak"].fillna(df["oldpeak"].mean())
```

```python
df.isnull().sum()
```

```python

df.dtypes
```

```python
df['num'] = df['num'].apply(lambda x: 1 if x > 0 else 0)
```

```python
print(df['sex'].unique())
print(df['exang'].unique())
```

```python
df["exang"] = df["exang"].replace({
    'False' : 'FALSE',
    'TURE' : 'TRUE',
    'True' : 'TRUE'
})
```

```python
df['exang'] = df['exang'].astype(str).str.lower().str.strip()
```

```python
#Label Encoding
df['sex'] = df['sex'].map({'Male':1, 'Female':0})
df['exang'] = df['exang'].map({'false':1, 'true':0})
```

```python
print(df['sex'].unique())
print(df['exang'].unique())
```

```python
#One hot coding
df = pd.get_dummies(df, columns=["cp", "fbs", "restecg","slope","thal"], drop_first=True)
```

```python
df.head(20)
```

```python
df.dtypes
```

```python
x = df.drop('num', axis =1)
y = df['num']
```

```python
print(type(df))
print(type(x))
```

```python
x.info()
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(y_train.unique)
print(y_train.dtype)
```

```python
#For better accuracy use RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(class_weight='balanced', random_state=42)
clf.fit(x_train, y_train)
```

```python
df['num'].value_counts()
```

```python
#Import model DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion="entropy", max_depth=7, random_state=42, min_samples_split=15, min_samples_leaf=10)
model.fit(x_train, y_train)
```

```python
#Generate prediction on Test data
y_pred = model.predict(x_test)
```

```python
x.columns.tolist()
```

```python
#Check Model Performance
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print("classification_report", classification_report(y_test, y_pred))
print("confusion_matrix", confusion_matrix(y_test, y_pred))
```

```python
#Save the model
import pickle
with open("model.pkl", "wb") as p:
    pickle.dump(model , p)
```
