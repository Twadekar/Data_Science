```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset
df=pd.read_csv("Zoo.csv")
```

```python
df.head()
```

```python
#Check null values
df.isnull().sum()
```

```python
#Check data types
df.dtypes
```

```python
df["animal name"].unique()
```

```python
#Convert categorical data into numerical format
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["animal name"]= le.fit_transform(df["animal name"])
le_name_mapping = dict(zip(le.classes_,le.transform(le.classes_)))
#print("features", animal name)
print("mapping", le_name_mapping)
```

```python
df.dtypes
```

```python
df.head()
```

```python
x= df[["hair","feathers","eggs","milk","airborne","aquatic","predator","toothed","backbone","breathes","venomous","fins","legs","tail","domestic","catsize"]]
y=df["type"]
```

```python
x.shape
```

```python
y.shape
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
y_test.value_counts()
```

```python
#create model for KNN
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=int(np.sqrt(len(x_train))), metric="euclidean")
```

```python
model.fit(x_train, y_train)
```

```python
#Generate prediction on Test data
y_pred = model.predict(x_test)
print(list(zip(y_test, y_pred)))
```

```python
feature_order = x.columns.tolist()
```

```python
#Check Model Performance
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cfm = confusion_matrix(y_test,y_pred)
print(cfm)

print("Classification report:")
print(classification_report(y_test,y_pred))

acc = accuracy_score(y_test,y_pred)
print("Accuracy of the Model: ",acc)
```

```python
#Save the Model
import pickle
with open("model.pkl", 'wb') as p:
    pickle.dump(model,p)
```
